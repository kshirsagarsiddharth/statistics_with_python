```python
import numpy as np 
import matplotlib.pyplot as plt 
```

## Example 1

```python
my_array = np.array([np.random.randint(1,6,100),np.random.randint(1,6,100),np.random.randint(1,6,100),np.random.randint(1,6,100)])
```

```python
plt.hist(my_array.mean(axis = 0))
```

#### population_std / sqrt(no_of_samples) = standard_deviation_of_sample_mean

```python
# this is population standard deviation
my_array.std() / np.sqrt(4)
```

```python
# This is standard error 
my_array.mean(axis = 0).std()
```

```python
def dice_rolls(a,b,n):
    return np.array([np.random.randint(a,b,n),np.random.randint(a,b,n),np.random.randint(a,b,n),np.random.randint(a,b,n)])
```

```python
plt.hist(dice_rolls(1,6,100).mean(axis = 0))
plt.show()
```

```python
plt.hist(dice_rolls(1,6,1000).mean(axis = 0))
plt.show()
```

```python
plt.hist(dice_rolls(1,6,10000).mean(axis = 0))
plt.show()
```

## Example 2

```python

sample_size = [1,10,100,1000,100000,1000000] 
means = [] 

for sample in sample_size: 
    # we are taking samples from -100,100 for 1000 times 
    means.append([np.mean(np.random.randint(-100,100,sample)) for _ in range(1000)]) 

```

```python
fig, ax = plt.subplots(nrows=3, ncols=2, figsize = (12,12)) 
ax = ax.ravel()

for i in range(len(sample_size)):
    ax[i].hist(means[i], density = True, color = 'green') 
    ax[i].set_title(f"Size = {sample_size[i]}")
```
